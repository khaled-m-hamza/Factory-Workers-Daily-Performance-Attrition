#Data Preprocessing

```python
#delete columns that not important for the model
data.drop(['sub_ID','sub_fname','sub_lname','sub_coll_IDs','sup_ID','sup_fname','sup_lname','event_date','recorded_note_from_sup'],axis='columns',inplace=True)
```

```python
#show dataset information after delete some columns
data.info()
```

```python
#convert non_numeric columns to numeric columns with Label encoder
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['sub_sex']=label_encoder.fit_transform(data['sub_sex'])
data['sub_shift']=label_encoder.fit_transform(data['sub_shift'])
data['sub_team']=label_encoder.fit_transform(data['sub_team'])
data['sub_role']=label_encoder.fit_transform(data['sub_role'])
data['sub_workstyle_h']=label_encoder.fit_transform(data['sub_workstyle_h'])
data['sup_sex']=label_encoder.fit_transform(data['sup_sex'])
data['sup_role']=label_encoder.fit_transform(data['sup_role'])
data['sub_workstyle_h']=label_encoder.fit_transform(data['sub_workstyle_h'])
data['event_weekday_name']=label_encoder.fit_transform(data['event_weekday_name'])
data['behav_comptype_h']=label_encoder.fit_transform(data['behav_comptype_h'])
data['behav_cause_h']=label_encoder.fit_transform(data['behav_cause_h'])
data['record_comptype']=label_encoder.fit_transform(data['record_comptype'])
data['record_cause']=label_encoder.fit_transform(data['record_cause'])
data['record_conf_matrix_h']=label_encoder.fit_transform(data['record_conf_matrix_h'])
```

```python
#show data
data.info()
```

```python
#convert data type of columns that have numeric values with data type object to int or float
data['sub_colls_same_sex_prtn'] = pd.to_numeric(data['sub_colls_same_sex_prtn'], errors='coerce')
data['sup_age'] = pd.to_numeric(data['sup_age'], errors='coerce')
data['sup_sub_age_diff'] = pd.to_numeric(data['sup_sub_age_diff'], errors='coerce')
data['sup_commitment_h'] = pd.to_numeric(data['sup_commitment_h'], errors='coerce')
data['sup_perceptiveness_h'] = pd.to_numeric(data['sup_perceptiveness_h'], errors='coerce')
data['sup_goodness_h'] = pd.to_numeric(data['sup_goodness_h'], errors='coerce')
```

```python
#show data information
data.info()
```

```python
#calumns that have value (None) replace it with 0 to make all calumn with numeric values
data['actual_efficacy_h'] = data['actual_efficacy_h'].replace('None', 0)

data['recorded_efficacy'] = data['recorded_efficacy'].replace('None', 0)
```

```python
#convert the data type of columns that have numeric values but with data type object
data['actual_efficacy_h'] = pd.to_numeric(data['actual_efficacy_h'], errors='coerce')
data['recorded_efficacy'] = pd.to_numeric(data['recorded_efficacy'], errors='coerce')
```

```python
#show data information
data.info()
```

```python
#check null values in columns
data.isnull().sum()
```

```python
#fill null values in column with mean value of the coulmn
data['sub_colls_same_sex_prtn'] = data['sub_colls_same_sex_prtn'].fillna(data['sub_colls_same_sex_prtn'].mean())
data['sup_age'] = data['sup_age'].fillna(data['sup_age'].mean())
data['sup_sub_age_diff'] = data['sup_sub_age_diff'].fillna(data['sup_sub_age_diff'].mean())
data['sup_commitment_h'] = data['sup_commitment_h'].fillna(data['sup_commitment_h'].mean())
data['sup_perceptiveness_h'] = data['sup_perceptiveness_h'].fillna(data['sup_perceptiveness_h'].mean())
data['sup_goodness_h'] = data['sup_goodness_h'].fillna(data['sup_goodness_h'].mean())
```

```python
#check null values in columns
data.isnull().sum()
```

```python
#show data information
data.info()
```

```python
#split data to X-->features and Y-->target
X = data.drop('record_conf_matrix_h', axis=1)  # Features
Y = data['record_conf_matrix_h']
```

```python
X=(X/X.max())
```

```python
#split data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='multinomial', max_iter=1000)
model.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Decode numerical predictions back to original labels
y_pred_decoded = label_encoder.inverse_transform(y_pred)
```

```python
#calculate precision ,recall ,f1_score and accuracy
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report
precision = precision_score(Y_test, y_pred, average='weighted')
recall = recall_score(Y_test, y_pred, average='weighted')
f1 = f1_score(Y_test, y_pred, average='weighted')
accuracy = accuracy_score(Y_test, y_pred)
```

```python
#print results
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-score: {f1:.4f}')
print(f'Accuracy: {accuracy:.4f}')
```
